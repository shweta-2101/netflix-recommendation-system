## Raw Data Loading Pipeline
Code below shows how raw data was processed — not rerun here due to
20GB file size. Processed CSV files are available in data/processed/

```python

# This code shows the pipeline used locally to generate ratings_sampled.csv
import pandas as pd
def load_ratings(filepath):
    rows = []
    movie_id = None
    with open(filepath, 'r') as f:
        for line in f:
            line = line.strip()
            if line.endswith(':'):
                movie_id = int(line[:-1])
            else:
                user_id, rating, date = line.split(',')
                rows.append([int(user_id), movie_id, int(rating), date])
    return pd.DataFrame(rows, columns=['user_id', 'movie_id', 'rating', 'date'])

# NOTE: This code was run locally on the full dataset
# combined_data_1-4.txt files are ~20GB total
# Output saved as ratings_sampled.csv
```
